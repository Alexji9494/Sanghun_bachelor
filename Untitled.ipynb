```python
samples = pd.read_csv('mouse_VISp_2018-06-14_samples-columns.csv')
exon_filtered_log2 = np.log2(pd.read_hdf('exon_filtered.h5', key='exon_filtered'))
exon_filtered_log2[exon_filtered_log2==exon_filtered_log2.values[0, 0]]=0
```

```python
L1_mean = np.mean(L1_log2, axis=1)
L23_mean = np.mean(L23_log2, axis=1)
L4_mean = np.mean(L4_log2, axis=1)
L5_mean = np.mean(L5_log2, axis=1)
L6_mean = np.mean(L6_log2, axis=1)
```
